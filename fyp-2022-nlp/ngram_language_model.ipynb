```python
# Import data\hate\test_text.txt txt file into dataframe
df_train = pd.read_csv('data/news-commentary-v16.en', sep='\t', header=None, nrows=5000)
# Import corresponding labels into dataframe
data = list(df_train[0])
data_tokens = [word_tokenize(i) for i in data]
```

```python
train, vocab = padded_everygram_pipeline(2, data_tokens)
text = open('data/news-commentary-v16.en', 'r')
train2 = ( ngram for sent in text for ngram in ngrams(sent, 2, pad_left = True, pad_right = True, right_pad_symbol='<s>', left_pad_symbol="</s>"))
```

```python
lm = MLE(2)
lm.fit(train, vocab)

```

```python
print(lm.score("a"))
print(lm.score("prices", ["Gold"]))
print(lm.vocab)
print(lm.counts)
print(lm.generate(30, random_seed=3))

```

```python
freq_dist = nltk.FreqDist(train2)
kneser_ney = nltk.KneserNeyProbDist(freq_dist)
prob_sum = 0
count = 0
for i in kneser_ney.samples():
    prob_sum += kneser_ney.prob(i)
    count += 1
print(prob_sum)
```
